```python
# @title **Experiment 5 - E. coli interactions predicted by human cross model**

import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, average_precision_score

trained_model_path = "human_xspecies3.1"
input_interactions_candidates = "all-data/data_cross_species/mouse_test.csv"
topology_features = "no"

!python reuse.py -trained_model_path {trained_model_path} -input_interactions_candidates {input_interactions_candidates} -topological_features {topology_features}

df = pd.read_csv(trained_model_path+'/folds_and_topology_feats/fold1/candidates_prediction.csv')
score= 'Candidate_score'
labels = df['Label']
probabilities = df[score]
auroc = roc_auc_score(labels, probabilities)
aupr = average_precision_score(labels, probabilities)
print()
print('Performance')
print(f"AUROC: {auroc}")
print(f"AUPR: {aupr}")
```
